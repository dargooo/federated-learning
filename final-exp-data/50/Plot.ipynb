```python
accs, times = [], []
with open('exp__async_hier__50__5__0.0005__0.01__8__0.9__100.txt', 'r') as f:
    line = f.readline()
    while line:
        ll = line.strip().split(',')
        if len(ll) == 3:
            accs.append(float(ll[1]))
            times.append(int(ll[0]))
        line = f.readline()
print(len(accs))
```
